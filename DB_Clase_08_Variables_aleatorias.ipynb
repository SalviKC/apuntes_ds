## 1. Variables aleatorias <a name="variables"></a>

<img src="https://drive.google.com/uc?export=view&id=1z9eni2W_ctTwxAWhUkE0otBfRG0RuveZ" height="150" width="520" style="float: center;">

- Una **variable aleatoria** (*random variable*) es una representaci√≥n num√©rica de un evento aleatorio.

- Asigna un n√∫mero a cada posible resultado de un experimento aleatorio.

- Debido a la naturaleza de los sucesos aleatorios, pueden ser discretas o continuas.

  **Variable aleatoria discreta** (*discrete random variable*): Puede tomar un n√∫mero finito contable de valores distintos.
  
  Por ejemplo, el n√∫mero de veces que aparece cara al lanzar una moneda 10 veces es una variable aleatoria discreta, ya que puede tener valores como 0, 1, 2, ..., 10.

  **Variable aleatoria continua** (*continuous random variable*): Puede tomar cualquier valor en un intervalo continuo.
  
  Por ejemplo, la altura de una persona seleccionada al azar es una variable aleatoria continua porque puede ser cualquier valor dentro de un rango razonable, como entre 50 cm y 250 cm.

- Cada variable aleatoria tiene asociada una **funci√≥n de distribuci√≥n** (*distribution function*), que describe la probabilidad de que la variable aleatoria tome un valor espec√≠fico (para variables discretas) o caiga dentro de un intervalo espec√≠fico (para las continuas).

### 1.1. Experimento aleatorio <a name="experimento"></a>

- Produce un **resultado impredecible**.
- A pesar de que los resultados individuales no se pueden predecir con certeza, el conjunto de todos los resultados posibles es bien definido.
- Los experimentos aleatorios son fundamentales en la teor√≠a de la probabilidad y se utilizan para modelar situaciones en las que hay incertidumbre.

**Caracter√≠sticas de un Experimento Aleatorio:**

**Impredecibilidad**: Los resultados individuales del experimento no se pueden predecir con certeza antes de que se realice el experimento.

**Espacio Muestral**: conjunto de todos los posibles resultados de un experimento aleatorio (denotado por ùëÜ o Œ©). Por ejemplo, en el lanzamiento de un dado, el espacio muestral es {1,2,3,4,5,6}.

**Repetibilidad**: Aunque los resultados individuales son impredecibles, el experimento se puede repetir bajo las mismas condiciones, y el comportamiento probabil√≠stico de los resultados se mantiene constante.

### Ejemplos de Experimentos Aleatorios:

- Lanzamiento de una moneda:
  
  El resultado puede ser cara (C) o cruz (X).
  
  El espacio muestral es {C,X}.

- Lanzamiento de un dado:

  El resultado puede ser cualquier n√∫mero entero del 1 al 6.
  
  El espacio muestral es {1,2,3,4,5,6}.

- Extracci√≥n de una carta de una baraja:

  El resultado puede ser cualquier carta de las 52 en una baraja est√°ndar.
  
  El espacio muestral incluye todas las cartas posibles, como {2‚ô°,3‚ô¢,‚Ä¶,K‚ô†,A‚ô£}.

## 2. Funciones de distribuci√≥n <a name="distribuci√≥n"></a>

Para variables aleatorias discretas representamos una distribuci√≥n a trav√©s de **funciones de masa de probabilidad** (*PMF*, *probability mass function*). La cual define la probabilidad de todos los valores posibles de la variable aleatoria.

Para variables aleatorias continuas representamos su distribuci√≥n por medio de la **funci√≥n de densidad de probabilidad** (*PDF*, *probability density function*).

#### Funci√≥n de masa de probabilidad

Es espec√≠fica para variables aleatorias discretas y da la probabilidad de que la variable aleatoria tome un valor concreto.

```python
data = np.arange(1, 7)
pmf = [1/6 for _ in data]

plt.figure(figsize = (5, 4))
plt.stem(data, pmf)
plt.title("PMF - Lanzar un dado")
plt.xlabel("Valor del dado")
plt.ylabel("Probabilidad")
plt.xticks(data)
plt.show()
```

**Ejemplo**: Lanzar un dado "justo" o "no cargado". La PMF para el resultado  "obtener un 3" ser√≠a P(X = 3), donde X es la variable aleatoria que representa el resultado del dado.

En este caso vemos que la distribuci√≥n de la variable aleatoria es **uniforme**, ya que sus probabilidades no cambian.

#### Funci√≥n de densidad de probabilidad

Es lo mismo que la PMF, pero para variables continuas.

```python
# 1000 valores entre -5 y 5
data = np.linspace(-5, 5, 1000)
pdf = norm.pdf(data)

plt.figure(figsize = (5, 4))
plt.plot(data, pdf, "r-")
plt.title("PDF - Distribuci√≥n normal")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.show()
```

**Ejemplo**: Medimos la altura de personas en una poblaci√≥n. La PDF nos dir√° la densidad de probabilidad de encontrar alguien con una altura espec√≠fica. No la probabilidad exacta de que alquien tenga una altura en particular.

En este caso vemos que la distribuci√≥n es **normal**, ya que el gr√°fico reproduce fielmente una campana gaussiana. Est√° determinada por dos par√°metros: la media y la desviaci√≥n est√°ndar.

#### Funci√≥n de distribuci√≥n acumulativa

Una **funci√≥n de distribuci√≥n acumulativa** (*CDF*, *cumulative distribution function*) est√° definida para variables discretas y continuas y mide la probabilidad de que la variable aleatoria tome un valor menor o igual a uno dado.

```python
data = np.linspace(-5, 5, 1000)
cdf = norm.cdf(data)

plt.figure(figsize = (5, 4))
plt.plot(data, cdf, "r-")
plt.title("CDF - Distribuci√≥n normal")
plt.xlabel("Valor")
plt.ylabel("Probabilidad")
plt.show()
```

- Se denota como F(x).
- Para una variable aleatoria ùëã, la CDF ùêπ(ùë•) = ùëÉ(ùëã‚â§ùë•) proporciona la probabilidad de que X tome un valor menor o igual a ùë•.
- La CDF es una funci√≥n no decreciente que va desde 0 a 1.
- Por ejemplo, si X es el resultado de lanzar un dado, F(3) ser√≠a la probabilidad de que el resultado sea 1, 2 o 3, lo que es 3 / 6 = 0.5.

## 3. Distribuciones de probabilidad <a name="distribuciones"></a>

Las **distribuciones de probabilidad** (*probability distributions*) describen c√≥mo est√°n distribuidas las probabilidades de una variable aleatoria.

Hay muchas distribuciones de probabilidad, y cada una describe un tipo diferente de proceso aleatorio.

#### Distribuci√≥n Binomial

Aplica a variables aleatorias discretas y describe el n√∫mero de **√©xitos** en un conjunto de **experimentos independientes de Bernoulli**.

Estos experimentos deben tener dos resultados posibles. Uno de las dos opciones.

Cada ensayo tiene dos posibles resultados: √©xito (con probabilidad
ùëù) o fracaso (con probabilidad 1‚àíùëù).

Por ejemplo, n√∫mero de caras obtenidas al lanzar una moneda 10 veces.

### Ejemplo de Distribuci√≥n Binomial

Supongamos que estamos lanzando una moneda 10 veces y queremos saber la probabilidad de obtener exactamente 4 caras (√©xitos).

- **n** (n√∫mero de ensayos): 10
- **p** (probabilidad de √©xito en cada ensayo): 0.5 (porque la probabilidad de obtener cara en un lanzamiento de moneda justa es 0.5)
- **k** (n√∫mero de √©xitos deseados): 4

La funci√≥n de masa de probabilidad de la distribuci√≥n binomial es:

$$
P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
$$

donde $\binom{n}{k}$ es el coeficiente binomial, calculado como:

$$
\binom{n}{k} = \frac{n!}{k!(n-k)!}
$$

Para este ejemplo, calculamos la probabilidad de obtener exactamente 4 caras en 10 lanzamientos de moneda:

$$
P(X = 4) = \binom{10}{4} (0.5)^4 (1 - 0.5)^{10 - 4}
$$

Calculamos el coeficiente binomial:

$$
\binom{10}{4} = \frac{10!}{4!(10-4)!} = \frac{10!}{4!6!} = 210
$$

Entonces, la probabilidad es:

$$
P(X = 4) = 210 \cdot (0.5)^4 \cdot (0.5)^6 = 210 \cdot (0.5)^{10} = 210 \cdot \frac{1}{1024} \approx 0.205
$$

Por lo tanto, la probabilidad de obtener exactamente 4 caras en 10 lanzamientos de una moneda justa es aproximadamente 0.205 (o 20.5%).

```python
3! = 3 * 2 * 1
5! = 5 * 4 * 3 * 2 * 1
```

```python
n_experiments = 10
probability = 0.5

data = range(n_experiments)
pmf = binom.pmf(data, n_experiments, probability)

plt.figure(figsize = (8, 4))
plt.stem(data, pmf)
plt.show()
```

#### Distribuci√≥n de Poisson

La **distribuci√≥n de Poisson** (*Poisson distribution*) aplica a variables aleatorias discretas y proporciona la probabilidad de que ocurra un determinado suceso un numero K de veces en un intervalo de tiempo.

Por ejemplo, n√∫mero de llamadas que recibe un call center en una hora.

## Distribuci√≥n de Poisson

$$
P(X = x) = \frac{e^{-\lambda} (\lambda)^x}{x!}, \, \lambda > 0, \, x = 0, 1, 2, 3, \dots
$$

Donde:

- **P(X=x)** : es la probabilidad de ocurrencia cuando la variable discreta $ X$ toma un valor finito $x$.
- **$\lambda$** : promedio de ocurrencias en un intervalo (tiempo, volumen, √°rea, etc.).
- **$e$** : tiene un valor aproximado de 2.71828183...
- **$x$** : es el n√∫mero de ocurrencias.


### Ejemplo de Distribuci√≥n de Poisson

Supongamos que un call center recibe en promedio 3 llamadas por minuto. Queremos calcular la probabilidad de recibir exactamente 5 llamadas en un minuto.

- **Œª** (tasa promedio de ocurrencia): 3 llamadas por minuto
- **k** (n√∫mero de eventos deseados): 5 llamadas

Para este ejemplo, calculamos la probabilidad de recibir exactamente 5 llamadas en un minuto:

$$
P(X = 5) = \frac{{3^5 e^{-3}}}{5!}
$$

Calculamos cada t√©rmino:

- $3^5 = 243$
- $e^{-3} \approx 0.0498$ (usando la aproximaci√≥n del valor de e)
- $5! = 120$

Entonces, la probabilidad es:

$$
P(X = 5) = \frac{{243 \cdot 0.0498}}{120} \approx \frac{{12.1054}}{120} \approx 0.1009
$$

Por lo tanto, la probabilidad de recibir exactamente 5 llamadas en un minuto es aproximadamente 0.1009 (o 10.09%).


```python
lambda_poisson = 3  # Tasa promedio de ocurrencia (Œª)
n_occur = 5  # N√∫mero m√°ximo de ocurrencias que queremos graficar

data = np.arange(0, n_occur * 2)  # Aumentamos el rango para incluir m√°s valores
pmf = poisson.pmf(data, lambda_poisson)

plt.figure(figsize=(8, 4))
plt.stem(data, pmf)
plt.title(f"Distribuci√≥n de Poisson (Œª = {lambda_poisson})")
plt.xlabel("N√∫mero de ocurrencias (k)")
plt.ylabel("P(X = k)")
plt.grid(True)
plt.show()
```

#### Distribuci√≥n geom√©trica

La **distribuci√≥n geom√©trica** (*geometric distribution*) aplica a variables aleatorias discretas y mide el n√∫mero de ensayos necesarios para obtener el primer √©xito en experimentos de Bernoulli independientes.

Por ejemplo, n√∫mero de lanzamientos de moneda necesarios para obtener la primera cara.

Sea $ X \sim \text{Geom√©trica}(p) $ la funci√≥n de probabilidad de la distribuci√≥n geom√©trica es:

$$
P[X = x] = (1 - p)^{x - 1} p
$$

### Ejemplo de Distribuci√≥n Geom√©trica

Supongamos que estamos lanzando una moneda y queremos calcular la probabilidad de que la primera cara (√©xito) aparezca en el tercer lanzamiento.

- \( p \) (probabilidad de √©xito en cada ensayo): 0.5 (porque la probabilidad de obtener cara en un lanzamiento de moneda justa es 0.5)
- \( k \) (n√∫mero de ensayos hasta el primer √©xito): 3

Calculamos la probabilidad de obtener la primera cara en el tercer lanzamiento:

$$
P(X = 3) = (1 - 0.5)^{3-1} \cdot 0.5
$$

Calculamos cada t√©rmino:

- $ (1 - 0.5) = 0.5 $
- $ 0.5^{3-1} = 0.5^2 = 0.25 $
- $ p = 0.5 $

Entonces, la probabilidad es:

$$
P(X = 3) = 0.25 \cdot 0.5 = 0.125
$$

Por lo tanto, la probabilidad de obtener la primera cara en el tercer lanzamiento es 0.125 (o 12.5%).


```python
n_experiments = 25
probability = 0.5

data = range(1, n_experiments + 1)  # Incluir el 25¬∫ experimento
pmf = geom.pmf(data, probability)

plt.figure(figsize=(8, 4))
plt.stem(data, pmf)
plt.title('Funci√≥n de Masa de Probabilidad (PMF) - Distribuci√≥n Geom√©trica')
plt.xlabel('N√∫mero de ensayos hasta el primer √©xito')
plt.ylabel('Probabilidad')
plt.grid(True)
plt.show()
```

#### Distribuci√≥n normal

La **distribuci√≥n normal** (*normal distribution*) aplica a variables aleatorias continuas y tiene forma de campana, por eso se suele llamar tambi√©n **distribuci√≥n Gaussiana**.

Est√° determinada por dos par√°metros: la media $\mu$ y la desviaci√≥n est√°ndar $\sigma$. Por ejemplo, la distribuci√≥n de alturas o pesos en una poblaci√≥n grande.

### Ejemplo distribuci√≥n Normal

Queremos calcular la probabilidad de que un hombre seleccionado al azar tenga una altura entre 170 cm y 180 cm. Con media 175 cm y desviaci√≥n est√°ndar de 10 cm.

La probabilidad de que una variable aleatoria $ X $ siga una distribuci√≥n normal $ N(\mu, \sigma^2) $ y tome un valor entre $ a $ y $ b $ se puede calcular usando la funci√≥n de distribuci√≥n acumulativa (CDF), que no tiene una f√≥rmula cerrada simple pero se puede calcular usando tablas de la normal est√°ndar o funciones de software.

1. **Convertir a la distribuci√≥n normal est√°ndar:**

$$
Z = \frac{X - \mu}{\sigma}
$$

2. **Calcular los valores \( Z \) para los l√≠mites:**

Para \( X = 170 \):

$$
Z_{170} = \frac{170 - 175}{10} = -0.5
$$

Para \( X = 180 \):

$$
Z_{180} = \frac{180 - 175}{10} = 0.5
$$

3. **Buscar las probabilidades correspondientes en la tabla de la distribuci√≥n normal est√°ndar:**

La probabilidad de que \( Z \) est√© entre $-0.5$ y $0.5$ se puede encontrar usando una tabla de la normal est√°ndar o funciones de software.

Usando una tabla de la normal est√°ndar, encontramos:

$$
P(Z \leq 0.5) \approx 0.6915
$$
$$
P(Z \leq -0.5) \approx 0.3085
$$

4. **Calcular la probabilidad:**

$$
P(170 \leq X \leq 180) = P(-0.5 \leq Z \leq 0.5)
$$

$$
P(-0.5 \leq Z \leq 0.5) = P(Z \leq 0.5) - P(Z \leq -0.5)
$$

$$
P(170 \leq X \leq 180) = 0.6915 - 0.3085 = 0.383
$$

Por lo tanto, la probabilidad de que un hombre seleccionado al azar tenga una altura entre 170 cm y 180 cm es aproximadamente 0.383 (o 38.3%).


```python
# Par√°metros de la distribuci√≥n normal
mu, sigma = 175, 10  # Media de 175 cm y desviaci√≥n est√°ndar de 10 cm

# Crear datos para graficar
data = np.linspace(140, 210, 1000)  # Rango de alturas amplio para ver la distribuci√≥n
pdf = norm.pdf(data, mu, sigma)

# Crear la figura
plt.figure(figsize=(8, 4))
plt.plot(data, pdf, label=f'N({mu}, {sigma}¬≤)', color='blue', lw=2)

# Sombrear el √°rea entre 170 y 180
x_fill = np.linspace(170, 180, 1000)  # Rango para el sombreado
plt.fill_between(x_fill, norm.pdf(x_fill, mu, sigma), color='orange', alpha=0.5, label='√Årea entre 170 cm y 180 cm')

# Agregar l√≠neas verticales para X = 170 y X = 180
plt.axvline(170, color='green', linestyle='--', label=f'X = 170 cm')
plt.axvline(180, color='orange', linestyle='--', label=f'X = 180 cm')

# Agregar t√≠tulo y etiquetas
plt.title('Distribuci√≥n Normal de Altura - N(175, 10¬≤)', fontsize=14)
plt.xlabel('Altura (cm)', fontsize=12)
plt.ylabel('Densidad de Probabilidad', fontsize=12)

# Agregar leyenda y cuadr√≠cula
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Mostrar la gr√°fica
plt.show()
```

#### Distribuci√≥n exponencial

La **distribuci√≥n exponencial** (*exponential distribution*) aplica a variables aleatorias continuas y describe el tiempo entre eventos en un proceso de Poisson.

Por ejemplo:
- El tiempo hasta que un dispositivo electr√≥nico falle. Es la contraparte continua de la distribuci√≥n geom√©trica.
- Tiempo de espera para ser atendido en un banco.
- Tiempo necesario para cargar un cami√≥n.

$$
P[T = t] = \lambda e^{-\lambda x}
$$

La funci√≥n de densidad de probabilidad (PDF) de la distribuci√≥n exponencial es:

$$
f(t; \lambda) = \lambda e^{-\lambda t}
$$

Y la funci√≥n de distribuci√≥n acumulativa (CDF) es:

$$
F(t; \lambda) = 1 - e^{-\lambda t}
$$

### Ejemplo de Distribuci√≥n Exponencial

Supongamos que estamos analizando el tiempo entre llamadas en un centro de atenci√≥n telef√≥nica, donde en promedio se recibe una llamada cada 5 minutos. Queremos calcular la probabilidad de que el tiempo hasta la pr√≥xima llamada sea menos de 3 minutos.

- **La tasa de ocurrencia**: $\lambda = \frac{1}{5}$ llamadas por minuto (ya que en promedio se recibe una llamada cada 5 minutos).
- **Tiempo hasta el pr√≥ximo evento (t)**: 3 minutos.

Para calcular la probabilidad de que el tiempo hasta la pr√≥xima llamada sea menos de 3 minutos, utilizamos la CDF:

$$
P(T < 3) = F(3; \lambda) = 1 - e^{-\lambda \cdot 3}
$$

Dado que $\lambda = \frac{1}{5}$:

$$
P(T < 3) = 1 - e^{-\frac{1}{5} \cdot 3}
$$

$$
P(T < 3) = 1 - e^{-0.6}
$$

Calculamos el valor de \(e^{-0.6}\):

$$
e^{-0.6} \approx 0.5488
$$

Entonces, la probabilidad es:

$$
P(T < 3) = 1 - 0.5488 \approx 0.4512
$$

Por lo tanto, la probabilidad de que el tiempo hasta la pr√≥xima llamada sea menos de 3 minutos es aproximadamente 0.4512 (o 45.12%).


```python
lambda_param = 1/5

tiempos = np.linspace(0, 5, 1000)  # Rango de valores para X
pdf = expon.pdf(tiempos, scale=1/lambda_param)  # Funci√≥n de densidad de probabilidad (PDF)

# Calcular el valor t correspondiente al 45% de probabilidad
p_target = 0.45
t_value = -np.log(1 - p_target) / lambda_param  # t correspondiente al 45% acumulado

# Crear la figura
plt.figure(figsize=(8, 4))
plt.plot(tiempos, pdf, label=f'Distribuci√≥n Exponencial ($\lambda$={lambda_param})', lw=2, color='blue')

# Rellenar el √°rea bajo la curva hasta el valor t correspondiente
t_fill = np.linspace(0, t_value, 1000)
pdf_fill = expon.pdf(t_fill, scale=1/lambda_param)
plt.fill_between(t_fill, pdf_fill, color='orange', alpha=0.6, label=f'√Årea = {p_target * 100}%')

# T√≠tulo y etiquetas
plt.title('Distribuci√≥n Exponencial - 45% de Probabilidad', fontsize=14)
plt.xlabel('t (Tiempo)', fontsize=12)
plt.ylabel('Densidad de Probabilidad', fontsize=12)

# Marcar el valor de t correspondiente al 45% de probabilidad
plt.axvline(x=t_value, color='red', linestyle='--', label=f't = {t_value:.2f}')

# Agregar leyenda y cuadr√≠cula
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Mostrar la gr√°fica
plt.show()
```

#### Distribuci√≥n uniforme

La **distribuci√≥n uniforme** (*uniform distribution*) aplica a variables aleatorias continuas y supone que todos los valores en un intervalo tienen la misma probabilidad de ocurrencia.

Por ejemplo, seleccionar un n√∫mero al azar entre 0 y 1.

Para una distribuci√≥n uniforme, la probabilidad de que una variable $ X $ est√© entre dos valores $ x_1 $ y $ x_2 $ se calcula como:

$$
P(x_1 \leq X \leq x_2) = \frac{x_2 - x_1}{b - a}
$$

Y la funci√≥n de probabilidad como:

$$
f(x) = \begin{cases}
\frac{1}{b - a}; & a \leq x \leq b \\
0 & \text{otros valores}
\end{cases}
$$

Donde ùëé y ùëè son los l√≠mites del intervalo.



### Ejemplo distribuci√≥n uniforme

Supongamos que estamos interesados en modelar el tiempo que un cliente tarda en ser atendido en una tienda. Sabemos que el tiempo de espera es siempre de entre 5 y 15 minutos, y cada valor dentro de este intervalo es igualmente probable (no hay preferencias por ning√∫n tiempo en particular).

Esto se puede modelar como una distribuci√≥n uniforme continua con los par√°metros: ùëé = 5 y ùëè =15.

La funci√≥n de densidad de probabilidad (f.d.p.) de esta distribuci√≥n ser√≠a:

$$
f(x) = \begin{cases}
\frac{1}{15 - 5} = \frac{1}{10}, & 5 \leq x \leq 15 \\
0 & \text{otros valores}
\end{cases}
$$

¬øCu√°n ser√° la probabilidad de que el cliente espere entre 7 y 10 minutos?

$$
P(7 \leq X \leq 10) = \frac{10 - 7}{15 - 5} = \frac{3}{10} = 0.3
$$

Por lo tanto, la probabilidad de que el cliente espere entre 7 y 10 minutos es del 30%.

```python
# Definir los par√°metros de la distribuci√≥n uniforme
a = 5  # l√≠mite inferior
b = 15  # l√≠mite superior

# Crear el rango de valores de x
x = np.linspace(0, 20, 1000)  # Valores de x desde 0 hasta 20

# Definir la funci√≥n de densidad de probabilidad usando scipy.stats.uniform
rv = uniform(loc=a, scale=b-a)
y = rv.pdf(x)

# Graficar la funci√≥n de densidad de probabilidad
plt.figure(figsize=(6, 4))
plt.plot(x, y, label=f'Uniforme({a},{b})', color='blue')

# Sombrear el √°rea correspondiente a P(7 ‚â§ X ‚â§ 10)
x_fill = np.linspace(7, 10, 100)
y_fill = rv.pdf(x_fill)
plt.fill_between(x_fill, y_fill, color='green', alpha=0.5, label='P(7 ‚â§ X ‚â§ 10)')

# Sombrear el √°rea del intervalo completo [5, 15]
plt.fill_between(x, y, where=((x >= a) & (x <= b)), color='skyblue', alpha=0.3)

# Configurar el gr√°fico
plt.title(f'Probabilidad de esperar entre 7 y 10 minutos')
plt.xlabel('x (Minutos)')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

#### Distribuci√≥n t de Estudiante

La **distribuci√≥n t de Estudiante** (*Student's t distribution*) aplica a variables aleatorias continuas y es similar a la distribuci√≥n normal pero con colas m√°s pesadas. Es √∫til cuando el tama√±o de la poblaci√≥n es reducido.

- Forma: La distribuci√≥n t de Student es sim√©trica y tiene forma de campana, similar a la distribuci√≥n normal. Sin embargo, tiene colas m√°s gruesas, lo que significa que tiene m√°s probabilidad de producir valores extremos.

- Grados de Libertad: La forma de la distribuci√≥n t depende de los grados de libertad (gl), que generalmente est√°n relacionados con el tama√±o de la muestra. Cuanto mayor sea el tama√±o de la muestra, m√°s se asemeja la distribuci√≥n t a la distribuci√≥n normal.

- Uso: Se utiliza principalmente en situaciones donde se estima la media de una poblaci√≥n con una muestra peque√±a (t√≠picamente n < 30) y la desviaci√≥n est√°ndar de la poblaci√≥n es desconocida.

La funci√≥n de densidad de probabilidad $ f(t) $ de la distribuci√≥n t de Student se define como:

$$
f(t) = \frac{\Gamma\left(\frac{v + 1}{2}\right)}{\sqrt{v \pi} \, \Gamma\left(\frac{v}{2}\right)} \left(1 + \frac{t^2}{v}\right)^{-\frac{v + 1}{2}}
$$

donde:
- $ t $ es el valor de la variable aleatoria,
- $ v $ son los grados de libertad,
- $ \Gamma $ es la funci√≥n gamma.

La funci√≥n de distribuci√≥n acumulativa $ F(t) $ se define como:

$$
F(t) = \int_{-\infty}^{t} f(x) \, dx
$$

No tiene una forma cerrada simple, pero se puede calcular num√©ricamente o consultar tablas de la distribuci√≥n t de Student.


```python
# Definir los grados de libertad
v = 10  # Puedes cambiar este valor para ver diferentes distribuciones

# Generar valores de t
x = np.linspace(-5, 5, 1000)

# Calcular la PDF
pdf = t.pdf(x, df=v)

# Graficar la PDF
plt.figure(figsize=(6, 4))
plt.plot(x, pdf, label=f'Distribuci√≥n t (v={v})', color='blue')
plt.title('Funci√≥n de Densidad de Probabilidad de la Distribuci√≥n t de Student')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='red', lw=0.5, ls='--', label='Media = 0')
plt.grid()
plt.legend()
plt.show()
```

#### Distribuci√≥n Chi-cuadrada

La **distribuci√≥n Chi-cuadrada** (*Chi-square distribution*) aplica a variables aleatorias continuas.

Nos indica si dos variables son independientes o no.

- Forma: La distribuci√≥n chi-cuadrada es asim√©trica y tiene una cola a la derecha. A medida que aumentan los grados de libertad, la distribuci√≥n se asemeja m√°s a una distribuci√≥n normal.

- Grados de Libertad: La forma de la distribuci√≥n depende de los grados de libertad (ùëò), que se relacionan con el n√∫mero de categor√≠as o el tama√±o de la muestra.

- Usos: Se utiliza en la prueba de hip√≥tesis para determinar si hay una diferencia significativa entre las frecuencias observadas y esperadas en una tabla de contingencia.

  Tambi√©n se usa en an√°lisis de varianza (ANOVA) y para evaluar la bondad de ajuste de modelos.

La funci√≥n de densidad de probabilidad \(\chi^2\) se define como:

$$
f(x; k) = \frac{1}{2^{k/2} \Gamma\left(\frac{k}{2}\right)} x^{\frac{k}{2}-1} e^{-\frac{x}{2}}
$$

donde:
- $ x $ es la variable aleatoria,
- $ k $ son los grados de libertad,
- $\Gamma$ es la funci√≥n gamma.
