## 1. ¿Algebra lineal? <a name="algebra"></a>

<img src="https://drive.google.com/uc?export=view&id=1kLUVNqVlwixQvWPrNDDH7WE9AEsVefdz" height="183" width="275" style="float: center;">

- Es un rama de las matemáticas que se centra en el estudio de los espacios vectoriales y transformaciones lineales.

<br>
<b>Espacio vectorial</b>

- También conocido como espacio lineal es una colección de vectores. Con los vectores se pueden realizar operaciones (como sumar o multiplicar entre sí).
- Los vectores representan diversas cosas como puntos en el espacio, fuerzas, velocidades, etc.

<br>
<b>Vector</b>

- Objeto que tiene una dimensión o longitud y dirección.
- Suelen representarse como listas ordenadas de números.
- Por ejemplo: (2,3) será un verctore de dimensión 2 y el vector (1,2,3) de dimensión 3.

<br>
<b>Transformaciones lineales</b>

- Es una función entre dos espacios vectoriales que preserva las operaciones de suma y multiplicaciones de vectores.
- Se suelen representar usando matrices.
- Por ejemplo: una transformación lineal puede ser una forma de mover puntos en un espacio. Tenemos un punto en un plano con coordenadas (x, y). Al multiplicar ambas coordenadas por 2 obtenemos un nuevo punto (2x, 2y). El nuevo punto está el doble de lejos del original. Es una transformación lineal llamada "escalamiento".
- En el ejemplo podemos pensar la transformación lineal como una receta para mover puntos en un espacio de manera ordenada y predecible. Cada punto se mueve siguiendo las mismas reglas, lo que permite transformar formas y patrones completos de una manera consistente.

<br>
<b>Matrices</b>

- Estructura formada por filas y columnas.
- Se utilizan para representar transformaciones lineales.
- Ejemplo: una matriz **A** puede transformar el vector **v** en un nuevo vector **w** mediante la multiplicación **Av = w**.

### 1.1 Aplicaciones <a name="aplicaciones"></a>

El álgebra lineal tiene muchas aplicaciones prácticas en diversas áreas, como:

- **Ciencias e ingeniería**: Modelado de sistemas físicos, análisis de estructuras, circuitos eléctricos.
- **Computación gráfica**: Representación y manipulación de imágenes y animaciones.
- **Economía y finanzas**: Modelos económicos, optimización de recursos.
- **Inteligencia artificial y aprendizaje automático**: Análisis de grandes conjuntos de datos, algoritmos de aprendizaje.

### 1.2 Aplicaciones en IA <a name="aplicacionesia"></a>

El álgebra lineal tiene un impacto directo sobre todos los procesos que involucren el tratamiento de datos:

- Los datos en un dataset están vectorizados. Las filas se insertan en un modelo una a la vez para realizar cálculos más fáciles y auténticos.
- Las imágenes tienen una estructura tabular (matriz). Las técnicas de edición de imágenes, como recortar y cambiar de tamaño, utilizan operaciones algebraicas.
- La regularización en ciertos algoritmos.
- El aprendizaje profundo funciona con vectores, matrices e incluso tensores.
- En la regresión lineal, se utiliza álgebra lineal para describir la relación entre las variables.
- Cuando encontramos datos irrelevantes, normalmente eliminamos las columnas redundantes, por lo que la técncia de análisis de componentes principales (PCA) usada actúa con la factorización de matrices.
- Con la ayuda del álgebra lineal, los sistemas de recomendación pueden tener datos más depurados.

## 6. Introducción a Derivadas <a name="matrices"></a>

### **Pendiente de una recta**

Medida de su inclinación, es decir, cuánto sube o baja la recta a medida que te mueves a lo largo de ella. Es una forma de cuantificar la "inclinación" de la recta en un gráfico.

  La pendiente (m) de una recta se define como el cambio en la coordenada y (la "altura") dividido por el cambio en la coordenada x (la "anchura") entre dos puntos cualesquiera de la recta.

```python
# Definimos los puntos
punto1 = (1, 2)  # (x1, y1)
punto2 = (3, 6)  # (x2, y2)

# Extraemos las coordenadas x e y de los puntos
x1, y1 = punto1
x2, y2 = punto2

# Calculamos los cambios en x e y
delta_y = y2 - y1
delta_x = x2 - x1

# Calculamos la pendiente
pendiente = delta_y / delta_x

# Imprimimos el resultado
print(f"La pendiente de la recta que pasa por los puntos {punto1} y {punto2} es {pendiente}.")
```

Esto significa que por cada unidad que avanzas en el eje x, la coordenada y aumenta en 2 unidades. La recta sube con una pendiente de 2.

```python
# Calculamos la intersección en y (y-intercept)
# Usamos la fórmula y = mx + b, donde b = y - mx para un punto dado (x, y)
interseccion = y1 - pendiente * x1

# Creamos un rango de valores x para graficar la recta
x_values = [x1 - 1, x2 + 1]  # Puedes ajustar estos valores según el rango que quieras mostrar
y_values = [pendiente * x + interseccion for x in x_values]

# Graficamos los puntos y la recta
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label=f'Recta: y = {pendiente:.2f}x + {interseccion:.2f}', color='blue')
plt.scatter([x1, x2], [y1, y2], color='red', zorder=5)
plt.text(x1, y1, f'({x1}, {y1})', fontsize=12, ha='right')
plt.text(x2, y2, f'({x2}, {y2})', fontsize=12, ha='right')

# Añadimos etiquetas y título
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de la recta que pasa por dos puntos')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```

- Pendiente positiva: Si la pendiente es positiva, la recta sube hacia la derecha. Cuanto mayor es el valor de la pendiente, más empinada es la subida.

- Pendiente negativa: Si la pendiente es negativa, la recta baja hacia la derecha. Cuanto más negativo es el valor de la pendiente, más empinada es la bajada.

- Pendiente cero: Si la pendiente es 0, la recta es horizontal y no sube ni baja.

- Pendiente indefinida: Si la recta es vertical, la pendiente es indefinida, ya que Δ𝑥=0 y no se puede dividir por cero.

### **Derivada**

- Herramienta para obtener información numérica de una gráfica (al igual que las integrales).

Ejemplo: movimiento de un objeto es una función que muestra como cambia la distancia a lo largo del tiempo se obtiene su velocidad en cualquier instante de tiempo. La velocidad es la derivada. (Las derivadas son esenciales en física).

- La derivada de una función en un punto específico nos dice cómo está cambiando el valor de la función en ese punto. Es como la "pendiente" de la función en ese punto.

En este ejemplo, usaremos la función f y su derivada:

$$
 f(x) = x^2  \ \   y   \ \   f'(x) = 2x
$$

```python
# Definir la función y su derivada
def funcion(x):
    return x**2

def derivada(x):
    return 2*x
```

```python
# Crear un rango de valores para x
x_valores = np.linspace(-10, 10, 400)

# Calcular los valores de la función y su derivada
y_valores = funcion(x_valores)
dy_valores = derivada(x_valores)

# Crear el gráfico
plt.figure(figsize=(10, 5))

plt.plot(x_valores, y_valores, label='f(x) = x^2', color='blue')
plt.plot(x_valores, dy_valores, label="f'(x) = 2x", color='red', linestyle='--')

# Añadir etiquetas y leyenda
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de la función y su derivada')
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

### Ejemplo: Maximizar ganancias

Encontrar el número optimo de productos que tengo que fabricar (y vender) para maximizar los beneficios de la empresa (ganancia).

```python
# generamos la ecuacion
cantidad = sy.symbols('cantidad')

# funcion de costo: costo de fabricar una unidad
# supongamos 50 costo fijo, 10 variable por cada unidad producida, costo "extra"
costo_total = 50 + 10*cantidad + 10*cantidad**2
costo_total
```

```python
# supongamos que se vende a 150
ingresos_totales = 150 * cantidad
ingresos_totales
```

```python
# rentabilidad
# nos da el resultado matemático haciendo la resta de las expresiones
ganancias = ingresos_totales - costo_total
ganancias
```

Queremos conocer la cantidad óptima de producto que debemos producir para maximizar el beneficio.

La derivada nos permite entender como va cambiando la función al cambiar la "cantidad" a producir. Cómo cambiarán los beneficioes en función de la cantidad a producir.

Buscamos cuando dejan de crecer los beneficios y encontrar el punto en donde producir más no trae mayores beneficios (por ejemplo se deben almacenar, se deterioran, etc). Donde la derivada es cero.

```python
# calculamos la derivada
derivada_ganancias = sy.diff(ganancias, cantidad)
derivada_ganancias
```

```python
# resolvemos la ecuación, cuando vale 0
# nos dará un mínimo o máximo
puntos_criticos = sy.solve(derivada_ganancias, cantidad)
puntos_criticos
```

Si fabricamos (y vendemos) 7 unidades se obtiene el máximo beneficio.

```python
# ¿cuál es el beneficio?
ganancia_maxima = ganancias.subs(cantidad, puntos_criticos[0])
ganancia_maxima
```

Al fabricar 7 la gananacia es de 440.

#### Gráficamente

```python
# Graficamente
valores_cantidad = np.linspace(0, 10, 100)
valores_cantidad
```

```python
# para cada valor de cantidad calcula el beneficio
valores_ganancias = [ganancias.subs(cantidad, valor) for valor in valores_cantidad]
valores_ganancias
```

```python
# para cada valor de ganancia calcula derivada
valores_derivadas = [derivada_ganancias.subs(cantidad, valor) for valor in valores_cantidad]
valores_derivadas
```

```python
plt.plot(valores_cantidad, valores_ganancias, label='Ganancias', color= 'b')
plt.plot(valores_cantidad, valores_derivadas, label='Derivada de ganancias', color= 'g')

plt.scatter([puntos_criticos[0]], [ganancia_maxima], label = 'Máxima ganancia', color = 'r')
plt.scatter([puntos_criticos[0]], [0], label = 'Derivada cero', color = 'orange')

plt.xlabel('Cantidad de unidades producidas')
plt.ylabel('Ganancias / Derivada de Ganancias')

plt.legend()
plt.show()
```

¿Cómo se relaciona con ML?

- Hacemos predicciones con un margen de error. Buscamos reducir el error. Buscamos el punto en que el acierto sea máximo.
- Vamos a tener funciones de "costo" o perdida: valor real - valor predicho (error).
- Queremos minimizar ese error.

## 7. Integrales <a name="integrales"></a>

<img src="https://drive.google.com/uc?export=view&id=1ZT6kUZYNWgARm-zF7plR2SpVNF6-1-ux" height="300" width="500" style="float: center;">

- Operación matemática contraria a las derivadas.
- La integral matemática es un concepto fundamental en cálculo que se usa para encontrar el área bajo una curva, entre otras cosas.
- Suma de Partes Pequeñas: La idea básica de una integral es sumar muchas partes pequeñas para obtener un total. Por ejemplo, si quieres encontrar el área de un terreno irregular, podrías dividirlo en muchos rectángulos muy delgados, calcular el área de cada rectángulo y luego sumarlas todas.


```python
# Definir la función
def f(x):
    return x

# Crear un rango de valores x
x = np.linspace(0, 2, 1000)
y = f(x)

# Graficar la función
plt.plot(x, y, label='f(x) = x')

# Rellenar el área bajo la curva
plt.fill_between(x, y, where=[(x_val >= 0 and x_val <= 2) for x_val in x], color='skyblue', alpha=0.4)

# Añadir etiquetas y leyenda
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Área bajo la curva f(x) = x de 0 a 2')
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

¿Cómo se relaciona con ML?

- Al medir algoritmos podemos utilizar el área bajo la curva para evaluar si sabe clasificar entre una clase u otra (clasificador binario).
- Un valor cercano a 1 indica un buen rendimiento del clasificador, donde logra separar bien las clases.
- Un valor cercano a 0.5 indica un rendimiento aleatorio, donde el clasificador no es mejor que una elección al azar.

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc

# Generar datos sintéticos
X, y = make_classification(n_samples=1000, n_features=20, random_state=42)

# Dividir los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

# Inicializar el clasificador (regresión logística)
classifier = LogisticRegression()

# Entrenar el clasificador
classifier.fit(X_train, y_train)

# Predecir probabilidades en el conjunto de prueba
probas = classifier.predict_proba(X_test)[:, 1]

# Calcular la curva ROC y el AUC
fpr, tpr, thresholds = roc_curve(y_test, probas)
roc_auc = auc(fpr, tpr)

# Graficar la curva ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

- Diagonal: Representa el rendimiento de un clasificador aleatorio.
- Un clasificador que predice al azar tendría una curva ROC que sigue esta línea.
- Curva ROC: muestra cómo varía la TPR (eje Y) conforme cambia la FPR (eje X) ajustando el umbral de decisión del clasificador.
- Idealmente, queremos que esta curva esté lo más alejada posible de la línea aleatoria (es decir, hacia la esquina superior izquierda del gráfico), lo que indica un mejor rendimiento del clasificador.
- Área bajo la Curva (AUC - Area Under the Curve): Es el área total bajo la curva ROC. Proporciona una medida cuantitativa del rendimiento global del clasificador.
- El punto más cercano al vértice superior izquierdo del gráfico representa el mejor rendimiento del clasificador (TPR alto con FPR bajo).
- Si la curva ROC cruza la línea aleatoria, esto indica que el clasificador no es mejor que una elección al azar para algunos umbrales de clasificación.
